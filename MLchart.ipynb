```
import matplotlib.pyplot as plt
import numpy as np
fig, ax=plt.subplots(1,1)
marks=np.array([20,60,25,30,35,50,60,55,51,81,98,70,90,45,65,76,87,98,34])
#ange=[0,20,40,60,80,100]
ax.hist(marks,histtype='step',color='red',range=(20,100))
ax.set_title('histogram of result')
ax.set_xlabel('marks')
ax.set_ylabel('no of student')
plt.show()
```

```
noofusers=[100,200,90,70,80]
pl=['java','python','c','c++','js']
c=['r','g','y','c','m']
plt.axis('equal')
plt.pie(noofusers,labels=pl,autopct='%.2f%%',shadow=True,colors=c,explode=[0.1,.1,0,0,0])
plt.legend(loc='upper right')
plt.show()
```
